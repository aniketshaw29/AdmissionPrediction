```python
#importing packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#read the csv files
df = pd.read_csv("admission.csv")
df
```

```python
X = df[['Gre', 'Toefl','Rating', 'Sop', 'Lor', 'Cgpa', 'Research']]
X
```

```python
Y = df[['Chance']]
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0) 
Linear_Reg = LinearRegression() #sklearn.linear_model.LinearRegression
Linear_Reg.fit(X_train, Y_train)
```

```python
#checking score   (same as above 2)
print("Train score:",Linear_Reg.score(X_train,Y_train)) 
print("Test score:",Linear_Reg.score(X_test,Y_test))
```

```python
pd.to_pickle(Linear_Reg,'admission_prediction.pickle') #creating a pickle file for future use of predicition algo
```
